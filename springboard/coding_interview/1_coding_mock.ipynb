This is a classification problem where we are going to use above CoverType Dataset.

Covertype Data Set is a dataset describing **cartographic features** of areas of land in the USA and also its forest type according to the US Forest Service.

The above dataset has 7 (1 to 7) classes but we are interested in only class **3**.

So our classification model **should be binary** that should classify whether the the **new data input is class 3 or not**.

In this test we use the Covertype Data Set, a dataset describing cartographic features of areas of land in the USA and also its forest type according to the US Forest Service. There are seven classes(COVER_TYPE) (1-7), 581012 samples and 54 features. For this test, we're only interested in cover type 3.

Note - COVER_TYPE is the last column

1. Build the model and evalute the accuracy

```python
# Import the relevant python libraries for the analysis
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
# Import data
covtype_data = pd.read_csv('covtype.csv')
covtype_df = DataFrame(covtype_data)
covtype_df.head()
```

```python
covtype_df.describe()
```

```python
covtype_df.info()
```

### Create a target (binary variable) that shows if a column has a '3' class in the 5 (Cover_Type) column
0 = TRUE, there is a 3 in the Cover_Type column

1 = FALSE, there is NOT a 3 in the Cover_Type column

```python
# Test output
covtype_df['5'][0].dtype
```

```python
# Create empty list
if_three = []

for classtype in covtype_df['5']:
   
    if classtype == 3:
        if_three.append(0)
    else:
        if_three.append(1)
```

```python
# Test output
#if_three
```

```python
# Add list to dataframe as target
covtype_df['target'] = pd.Series(if_three, np.arange(0,581011))
covtype_df.tail()
```

## Since the target variable is binary in nature, a logistic (probabalistic) regression model is ideal, to do it's sigmoidal mathmatical foundation. 

```python
# Separate out the features
X_features = covtype_df.copy().drop(columns='target')

# Separate out the target
y_target = covtype_df['target'].values
print(y_target.shape)
```

#### Testing and Training Data

```python
# Import necessary modules 
from sklearn.model_selection import train_test_split
```

```python
X_feature_train, X_feature_test, y_target_train, y_target_test = train_test_split(X_features, y_target, test_size=0.3, random_state=42)
len(X_feature_train), len(X_feature_test), len(y_target_train), len(y_target_test)
```

#### Standardize the Feature Data

```python
# Import necessary modules
from sklearn.preprocessing import StandardScaler
```

```python
# Create a scaler instance
scaler = StandardScaler()
```

```python
# Fit on training features set *only*
scaler.fit(X_feature_train)

# Apply transform to both the feature training and the test sets
X_train_scaled = scaler.transform(X_feature_train)
X_test_scaled = scaler.transform(X_feature_test)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Create a Logistic Regression Incidence 
clf_log = LogisticRegression(random_state=42, solver='lbfgs', n_jobs=10)
```

```python
# Fit data to model
clf_log_scaled = clf_log.fit(X_train_scaled, y_target_train)
```

#### Assess Accuracy of Model

```python
# Import necessary modules
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt 
```

```python
# Calculate the class probablity -returns the probability of model's ability to predict a likely value of the features
probability = clf_log_scaled.predict_proba(X_test_scaled)
probability
```

```python
# Predict the model -assess how well the model predicts unseen data
predict = clf_log_scaled.predict(X_test_scaled)

# Compute classification report (https://en.wikipedia.org/wiki/Precision_and_recall)
class_report = classification_report(y_target_test, predict)
print(class_report)
```

*Observations*: The model appears to have an overall accuracy of 75% when predicting '0', where a classtype IS 3, and 98% accuracy when predicting '1', where the classtype IS NOT 3.  

```python
# Print the accuracy from the testing data
'The accuracy score the the scaled data is: ', accuracy_score(y_target_test, predict)
```

#### The overall accuracy of this model is ~97%.
